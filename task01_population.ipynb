```python
import pandas as pd

df = pd.read_csv("data.csv", skiprows=4)
df.head()
```

```python
df.info()
```

```python
df.columns
```

```python
import matplotlib.pyplot as plt

year = "2022"

# choose real countries manually
real_countries = [
    "India", "China", "United States", "Indonesia",
    "Pakistan", "Brazil", "Nigeria", "Bangladesh",
    "Russia", "Mexico"
]

filtered = df[df["Country Name"].isin(real_countries)]

# plot
plt.figure(figsize=(10,6))
plt.bar(filtered["Country Name"], filtered[year])

plt.title("Top Countries by Population (2022)")
plt.xlabel("Country")
plt.ylabel("Population")
plt.xticks(rotation=45)

plt.show()
```

```python
filtered = filtered.sort_values(by=year, ascending=False)

plt.figure(figsize=(10,6))
plt.bar(filtered["Country Name"], filtered[year])
plt.title("Top Countries by Population (2022)")
plt.xlabel("Country")
plt.ylabel("Population")
plt.xticks(rotation=45)
plt.savefig("population_chart.png")
plt.show()
```
